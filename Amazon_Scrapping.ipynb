```python
import requests
from bs4 import BeautifulSoup
import time
import random
import pandas as pd
```

```python
BASE_URL = "https://www.snapdeal.com/search?keyword={}&page={}"
```

```python
def scrape_snapdeal_products(search_query, max_pages=5):
    products = []

    for page in range(1, max_pages + 1):
        url = BASE_URL.format(search_query, page)
        headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.99 Safari/537.36"
        }

        response = requests.get(url, headers=headers)

        if response.status_code != 200:
            print(f"Failed to retrieve page {page}")
            break

        soup = BeautifulSoup(response.text, "html.parser")
        product_containers = soup.find_all("div", class_="product-tuple-description")

        if not product_containers:
            print(f"No products found on page {page}")
            break

        for product in product_containers:
            try:
                name = product.find("p", class_="product-title").text.strip()
                price = product.find("span", class_="lfloat product-price").text.strip()
                link = product.find("a")["href"]

                products.append({"Name": name, "Price": price, "URL": link})
            except AttributeError:
                continue  # Skip if any data is missing

        print(f"Scraped page {page} successfully!")
        time.sleep(2)  # Delay to prevent getting blocked

    return products
```

```python
if __name__ == "__main__":
    search_query = input("Enter product name to search: ").replace(" ", "+")
    max_pages = int(input("Enter number of pages to scrape: "))

    data = scrape_snapdeal_products(search_query, max_pages)

    if data:
        df = pd.DataFrame(data)
        df.to_csv("snapdeal_products.csv", index=False)
        print("Data saved to snapdeal_products.csv")
    else:
        print("No data scraped.")
```
